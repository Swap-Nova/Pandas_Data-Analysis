<h5> Importing Pandas </h5>

```python
import pandas as pd
```

<h5> Creating an series: One-dimensional Object </h5>

```python
cars_series = pd.Series(["BMW", "Honda", "Audi"])
cars_series
```

```python
car_color_series = pd.Series(["Red", "White", "Black"])
car_color_series
```

<h5> Creating an Dataframe: Two-dimensional object (rows-columns) in the form of dict </h5>

```python
car_data = pd.DataFrame({"Car Company": cars_series, "Car Color": car_color_series})
car_data
```

<h5> Importing Dataset instead of creating an dataframe </h5>

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

<h5> Exporting the dataframe created in csv form </h5>

```python
car_sales.to_csv("exported-car-sales.csv")
```

```python
# Again reading the csv content exported
exported_car_sales = pd.read_csv("exported-car-sales.csv")
exported_car_sales
```

<p>But here we an can see that the columns that contains car model names is Unnamed here. Therefore, while exporting the csv file we have to make sure that the index is false.</p>

```python
# Point to remember here is that 'False' is case-sensitive 
car_sales.to_csv("alternate-exported-car-sales.csv", index=False)
```

```python
# Again regarding the csv content exported
alternate_exported_car_sales = pd.read_csv("alternate-exported-car-sales.csv")
alternate_exported_car_sales
```
